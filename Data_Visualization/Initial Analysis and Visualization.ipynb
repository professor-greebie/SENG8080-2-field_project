```python
import pandas as pd
import numpy as np
import pyranges as pr
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Importing the dataset
gr = pr.read_gtf('D:\DataFile\dipOrd1.ensGene.gtf')
df = gr.df
```

```python
# Number of unique values in each column
print(df.nunique())
```

```python
# Finding the number of null values that exist in the columns of the dataset
print(df.isnull().sum())

#The missing numbers could reflect gaps in the annotation process or limits in the genetic data available. It's probable that 
#not all transcript exons in the dataset were annotated or recognised throughout the sequencing and assembly process.
```

```python
# Counting unique values in each column
unique_chromosomes = df['Chromosome'].nunique()
unique_features = df['Feature'].nunique()
unique_strands = df['Strand'].nunique()

# Printing the results
print("Number of unique chromosomes:", unique_chromosomes) # There are a total of 25,760 unique chromosomes 
print("Number of unique features:", unique_features) # There are a total of 5 unique features 
print("Number of unique strands:", unique_strands) # There are a total of 2 unique strands 
```

```python
# Grouping by 'gene_name' and counting the occurrences of each gene
gene_transcript_count = df.groupby('gene_name')['transcript_id'].nunique()
gene_exon_count = df.groupby('gene_name')['exon_id'].nunique()

# Finding the gene with the most transcripts
gene_with_most_transcripts = gene_transcript_count.idxmax()

# Finding the gene with the most exons
gene_with_most_exons = gene_exon_count.idxmax()

# Printing the results
print("Gene with the most transcripts:", gene_with_most_transcripts)
print("Gene with the most exons:", gene_with_most_exons)
```

```python
# Group by 'exon_id' and 'transcript_id' to find transcripts sharing the same exons
transcripts_sharing_exons = df.groupby(['exon_id'])['transcript_id'].apply(lambda x: ','.join(x.dropna().unique()))

# Print the result
print("Transcripts sharing exons:")
print(transcripts_sharing_exons)

'''
The result includes 'exon_id' values as well as their associated 'transcript_id' values, suggesting that certain transcripts 
share exons in the genomic data. For example, 'ENSDORT00000000002.1.1' is a 'exon_id' that is linked to 'ENSDORT00000000002.1', 
which is a transcript that shares this exon. Similarly, 'ENSDORT00000023118.1.1' is another 'exon_id' that shares the exon with 
its matching 'ENSDORT00000023118.1' transcript. The output displays a list of 'exon_id' and the transcripts that share each exon,
providing insights into the genomic data's interactions between exons and transcripts.
'''
```

```python
feature_distribution = df['Feature'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(feature_distribution, labels=feature_distribution.index, autopct='%1.1f%%', startangle=90)
plt.title('Feature Distribution')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
# Group the data by 'Strand' and 'Feature' columns and count occurrences
strand_feature_counts = df.groupby(['Strand', 'Feature']).size().unstack(fill_value=0)

# Visualize the strand-specific feature counts using a stacked bar plot
plt.figure(figsize=(8, 6))
strand_feature_counts.plot(kind='bar', stacked=True)

# Set plot labels and title
plt.xlabel('Strand')
plt.ylabel('Count')
plt.title('Strand-Specific Feature Counts')
plt.legend(title='Feature')

# Show the plot
plt.tight_layout()
plt.show()
```
